# Oscillators phase space

```python
import numpy as np
import matplotlib.pyplot as plt
import RungeKutta as RK4
```

## Lorentz Oscillator Solution

```python
sigma = 10.0 #Variable for dx/dt
rho = 28.0 #Variable for dy/dt
beta = 8/3 #Variable for dz/dt
t = 0 #Starting time
tf = 30 #Ending time
h = 0.01 #Step size for RK4
#These variables were used since they can be easily found for comparison, 
#for example, in Wikipedia under the Lorenz system article
```

## Movement equations

$\frac{dx}{dt} = \sigma (y-x)$

$\frac{dy}{dt} = x(\rho-z)-y$

$\frac{dz}{dt} = (yx-\beta z)$

```python
#Derivative function to work with RK4 loop
def derivative(r,t,params):
    x = r[0]
    y = r[1]
    z = r[2]
    return np.array([sigma * (y - x), x * (rho - z) - y, (x * y) - (beta * z)])
```

```python
time = np.array([]) #Empty time array to fill for the x-axis
x = np.array([]) #Empty array for x values
y = np.array([]) #Empty array for y values
z = np.array([]) #Empty array for z valuesr = np.array([1.0, 1.0, 1.0]) #Initial conditions array
r = np.array([1.0, 1.0, 1.0]) #Initial conditions array
```

```python
while (t <= tf ):        #Appending values to graph
        time = np.append(time, t)
        z = np.append(z, r[2])
        y = np.append(y, r[1])
        x = np.append(x, r[0])        #RK4 Step method
        r = RK4.RKN4_1fstODE(derivative,r,t,h,[])        #Updating time value with step size
        t = t + h
```

```python
#Multiple graph plotting
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (15, 5))
ax1.plot(x, y)
ax1.set_title("X & Y")
ax2.plot(x, z)
ax2.set_title("X & Z")
ax3.plot(y, z)
ax3.set_title("Y & Z")
plt.show()
```

## Harmonic Oscillator

```python
h = 0.01
time = np.arange(0,10+h,h) 
```

```python
def HO_Acceleration(x,v,t,w):
    ddx = -(w*x)
    return ddx
```

```python
X = np.zeros(len(time),dtype='float') #Empty array for x values
V = np.zeros(len(time),dtype='float') #Empty array for x values
X[0] = 1
V[0] = 0
```

```python
for n in range(0,len(time)-1):#Appending values to graph
    X[n+1], V[n+1] = RK4.RKN4_2ndODE(HO_Acceleration,X[n],V[n],time[n],h,2)        #Updating time value with step size
           #RK4 Step method
```

```python
%matplotlib qt
```

```python
plt.figure(figsize = (12, 8))
plt.plot(time,X, '-', label=r'$v_{0} = 0, x_{0} = 1$')
plt.plot(time,np.cos(2*time), '--', label=r'$\cos(2t), v_{0}=0, x_{0}=1$')
# plt.plot(time,np.sin(2*time)/2, '--', label=r'$v_{0}=1, x_{0}=0$')
plt.title('Harmonic OScillator, Approximate Solution')
plt.xlabel(r'$t$')
plt.ylabel(r'$x(t)$')
plt.grid()
#lt.yscale('log')
# plt.ylim(-5,20)
plt.legend(loc='lower right')
plt.show()
```
